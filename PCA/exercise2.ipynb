```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
data = {
    'Variable': ['Climate', 'Housing', 'Health', 'Crime', 'Transportation',
                 'Education', 'Arts', 'Recreation', 'Economy'],
    'PC1': [0.90, 0.844, 0.786, 0.263, 0.406, 0.875, 0.945, 0.903, 0.142],
    'PC2': [0.017, 0.004, 0.234, 0.805, 0.420, 0.134, 0.115, 0.060, 0.150],
    'PC3': [0.097, 0.084, 0.036, 0.204, 0.364, 0.123, 0.021, 0.037, 0.239]
}
```

```python
df = pd.DataFrame(data)
```

```python
df.set_index('Variable', inplace=True)
```

```python
print("PCA Component Loadings Table:")
print(df)
```

```python
print("\nMost Influential Variable for Each Principal Component:")
for component in df.columns:
    max_variable = df[component].idxmax()
    max_value = df[component].max()
    print(f"{component}: {max_variable} (Loading: {max_value})")
```

```python
df.plot(kind='bar', figsize=(12,6), title='Variable Loadings on Principal Components')
plt.ylabel("Loading")
plt.xlabel("Variable")
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
from mpl_toolkits.mplot3d import Axes3D
```

```python
fig = plt.figure(figsize=(10, 7))
ax = fig.add_subplot(111, projection='3d')
```

```python
xs = df['PC1']
ys = df['PC2']
zs = df['PC3']
```

```python
ax.scatter(xs, ys, zs, s=100, c='r', depthshade=True)

for var, x, y, z in zip(df.index, xs, ys, zs):
    ax.text(x, y, z, var, fontsize=9)

ax.set_xlabel('PC1')
ax.set_ylabel('PC2')
ax.set_zlabel('PC3')
ax.set_title("3D Visualization of Variable Loadings")

plt.show()
```
